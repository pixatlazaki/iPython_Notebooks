```python
import scipy.integrate as integrate
import matplotlib.pyplot as plt
import numpy as np
```

```python
pi = np.pi
sqrt = np.sqrt
cos = np.cos
sin = np.sin
exp = np.exp
```

```python
beta = .7 # dampening param.
omega_0 = .5 # natural freq.
gamma = 1.1 # drive strength multiplier
omega = .5 # drive freq.
```

```python
def deriv_z(z, t):
    u, udot = z
    return [udot, -2*beta*udot + gamma*(omega_0**2)*cos(omega*t) - (omega_0**2)*sin(u)]
```

```python
t = np.linspace(0, 2.0*pi, 10000)
zinit = [0, 0]
z = integrate.odeint(deriv_z, zinit, t)
phi, phidot = z.T
plt.plot(t,phi,'b',t,phidot,'g')
```

```python
omega_0 = 9.1 * omega
beta = .5 * beta

t = np.linspace(0, 7.0*pi, 5000)
zinit = [1.49907, 0]
z = integrate.odeint(deriv_z, zinit, t)
phi, phidot = z.T
plt.plot(t,phi)
plt.show()
```
